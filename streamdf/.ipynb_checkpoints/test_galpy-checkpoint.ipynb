# Now we're gonna do some streamdf stuff

```python
from galpy.df import streamdf
from galpy.orbit import Orbit
from galpy.potential import LogarithmicHaloPotential
from galpy.actionAngle import actionAngleIsochroneApprox
from galpy.util import conversion #for unit conversions
```

These packages allow us to compute the position of the stream based off a fixed Milky Way potential:

```python
lp= LogarithmicHaloPotential(normalize=1.,q=0.9)
aAI= actionAngleIsochroneApprox(pot=lp,b=0.8)
```

LogarithmicHaloPotential is a function that sets up the potential of the Milky Way:
- Other potential models can be received from the galpy library
- Syntax: galpy.potential.\<nameofmodelpotential>
- The parameters of said model potential can be set in the function of its argument
    all specific models can be found here: https://docs.galpy.org/en/latest/reference/potential.html
    this is an example potential model, with arguments that specify the quantitative properties of said model (such as amp, normalize, q = potential flattening, etc.) https://docs.galpy.org/en/latest/reference/potentialloghalo.html 
- In this example, we directly import the LogarithmicHaloPotential 

actionAngleIsochroneApprox is a function that:
- approximates the given potential as an isochrone, 
- uses this approx to calculate the action of a given orbit (the approx makes this analytically sufficient apparently)
- as well as a fit for the angles to derive the angles
arguments for this function include the potential you are using that it will approximate (pot=) and the scale radius of the isocrone potential (b=). Additional arguments can be found here: https://docs.galpy.org/en/latest/reference/aaisochroneapprox.html 

```python
obs= Orbit([1.56148083,0.35081535,-1.15481504,0.88719443,-0.47713334,0.12019596])
```

This code essentially starts the orbit with the stated initial conditions. These initial conditions can be implemented in several ways, such as using astropy's SkyCoord and also just inputing a list of initial values as in this example. (It is probably best practice to define a list of initial conditions rather than to hard code it.) Likely would need some semblance of the position of Leo II, whether that is RA,dec or otherwise. https://docs.galpy.org/en/v1.8.3/reference/orbitinit.html 

```python
sigv= 0.365 #km/s
sdf= streamdf(sigv/220.,progenitor=obs,pot=lp,aA=aAI,leading=True,nTrackChunks=11,tdisrupt=4.5/conversion.time_in_Gyr(220.,8.))
sdf.plotCompareTrackAAModel()
```

streamdf sets up the simulation. The arguments for the function were the results of our previous lines of code:
- sigv is just the radial velocity dispersion of the progenitor
- progenitor takes the orbit object calculated earlier from initial conditions
- pot is just potential defined much earlier
- aA is the actionAngleIsochrone object set up earlier. Like before, it approximates the action through an isochrone potential
- leading: whether we model the leading or trailing part of the stream (True = model leading)
- nTrackIterations: number of iterations to perform along the track (I assume similar to a simulation's resolution)

This returns the natural log of the probability distribution of the phase space. That is, it computes the likelihood of a particle to be within a particular region of phase space and takes the ln of it. The arguments imputed are from the orbit object set up previously, basically setting it up such that it gives the likelihood of finding the star trail at the location of the progenitor. 

```python
sdf(obs.R()-0.1,obs.vR(),obs.vT(),obs.z()+0.1,obs.vz(),obs.phi())
# array([-inf])
# moving the range away too far from orbits produces a 0 probability. 
numpy.log(0)
```

```python
meanp, varp= sdf.gaussApprox([None,None,2./8.,None,None,None])
```

Approximating the distribution as Gaussian, this function returns the mean and variance (square of standard deviation). lb returns these values in observable units. 

```python
meanp[0]*8.
# 14.267559400127833
```

```python
numpy.sqrt(varp[0,0])*8.
# 0.04152968631186698
```

```python
xs= numpy.linspace(-3.*numpy.sqrt(varp[0,0]),3.*numpy.sqrt(varp[0,0]),21)+meanp[0]
logps= numpy.array([sdf.callMarg([x,None,2./8.,None,None,None]) for x in xs])
ps= numpy.exp(logps)
```

This is where we model the probability with respect to X. This block generates an X axis that spans 3 sigma from the mean of the distribution. This in effect is done because we know the features of a Gaussian/Normal distribution: the probability density produces a bellcurve such that its maximum is at the mean (which is typically centered to 0) and the area under the curve indicates the probability of falling within those x values. There is a 68% chance to be within 1 sigma, or standard deviation, from the mean, a 95% chance to be within 2 sigma, and a 99.5% chance to fall within 3 sigma. (These values can be calculated by integraing a normalized gaussian distribution across said sigma intervals--I imagine that it is not easy). 

sdf. callMarg calculated the pdf in much of the same way but marginalizes in some directions (this is necessary as we are trying to calculated p(x|z), or the probability of the star being at x given some z value). Since this returns the natural log of the probability, the proper value of the probability is derived by raising the output of the previous function as a power of e. 

```python
ps/= numpy.sum(ps)*(xs[1]-xs[0])*8.
```

This simply normalizes our distribution. Specifically, probabilities must fulfill the property that the sum of all probabilities must equal 1 (as there is a 100% chance that SOMETHING will happen). With a probability density in continuous space, this manifests as an integral, as integrating over the entire phase space will produce a 100% probability in finding our point/star trail. Thus, we must normalize a probability density that does not fulfill this quality by dividing it by its total integral (this is because this forces the function to fulfill the normalization condition: int(f(x)/int(f(x))) = 1). 

In the code, this is further multiplied by the spacing between xs (dx) and 8. To be honest, I'm not sure why, but it probably has something to do with the units, as each term of an integral is multiplied by a dx and 8 just converts dx into standard units. 

```python
plt.plot(xs*8.,ps)
plt.plot(xs*8.,1./numpy.sqrt(2.*numpy.pi)/numpy.sqrt(varp[0,0])/8.*numpy.exp(-0.5*(xs-meanp[0])**2./varp[0,0]))
```

Looking at the star trail, we can see that there are 2 values of Z where the star trail is for values of X greater than 13 kpc. Thus, to calculate p(Z|X), the probability of the star trail to be at Z for a given value X, it will be in our best interest to highlight a certain region of phase space and call a function to find the closest star trail. 

```python
cindx= sdf.find_closest_trackpoint(13.5/8.,None,5.32/8.,None,None,None,xy=True)
```

```python
meanp, varp= meanp, varp= sdf.gaussApprox([13.5/8.,None,None,None,None,None],cindx=cindx)
```

```python
meanp[1]*8.
# 5.4005530328542077
```

```python
numpy.sqrt(varp[1,1])*8.
# 0.05796023309510244
```

We approximate this as a Gaussian as well. Thus, we can do the same trick as previously. 

```python
zs= numpy.linspace(-3.*numpy.sqrt(varp[1,1]),3.*numpy.sqrt(varp[1,1]),21)+meanp[1]
logps= numpy.array([sdf.callMarg([13.5/8.,None,z,None,None,None],cindx=cindx) for z in zs])
ps= numpy.exp(logps)
ps/= numpy.sum(ps)*(zs[1]-zs[0])*8.
```

```python
plt.plot(zs*8.,ps)
plt.plot(zs*8.,1./numpy.sqrt(2.*numpy.pi)/numpy.sqrt(varp[1,1])/8.*numpy.exp(-0.5*(zs-meanp[1])**2./varp[1,1]))
```
