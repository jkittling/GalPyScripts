```python
#Galpy imports
import galpy
from galpy.orbit import Orbit
import galpy.potential as p
from galpy.potential import MWPotential2014 as MWPotential
from galpy.util import coords, conversion #for unit conversions

#Other imports
import numpy as np
import matplotlib.pyplot as plt
import astropy.coordinates as coord
from astropy import units as u
```

```python
def astro_to_galcen(ra, dec, dist):
    """
    takes in icrs coords and convert them
    to galactocentric coordinates
    """
    cicrs = coord.SkyCoord(ra=ra*u.degree, dec=dec*u.degree, distance=dist*u.kpc,
                       frame="icrs")
    
    cg = cicrs.transform_to(coord.Galactocentric)
    
    return np.array([cg.x.value, cg.y.value, cg.z.value])

def astro_to_icrs(x, y, z):
    """
    takes in galactocentric coords and convert them
    to icrs coordinates
    """
    cg = coord.SkyCoord(x=x*u.kpc, y=y*u.kpc, z=z*u.kpc,
                       frame=coord.Galactocentric)
    
    cicrs = cg.transform_to(coord.ICRS)
    
    return np.array([cicrs.ra.value, cicrs.dec.value, cicrs.distance.value])

def galpy_coord(orb, i, sys):
    """
    takes in icrs coords and convert them
    to galactocentric coordinates. sys 
    argument indicates which coordinate
    system to convert to. 
    """
    if sys == "galcen":
        ###GALPY IS LEFT HANDED! 
        ###Changed the sign of x axis to make frame right-handed!
        return np.array([-orb.x(t[i],),orb.y(t[i]),orb.z(t[i])])
    if sys == "icrs":
        return np.array([orb.ra(t[i]),orb.dec(t[i]),orb.dist(t[i])])
    
def plot(axs, r, t, sys, color):
    """
    takes a 3D axis object and the 
    position vectors to plot it over time 
    """
    
    axs[0].plot(t, r[:,0], color = color)
    axs[1].plot(t, r[:,1], color = color)
    axs[2].plot(t, r[:,2], color = color)
    
    for ax in axs:
        ax.set_xlabel("Time")
    if sys == "galcen":
        axs[0].set_ylabel("X")
        axs[1].set_ylabel("Y")
        axs[2].set_ylabel("Z")
    if sys == "icrs":
        axs[0].set_ylabel("ra")
        axs[1].set_ylabel("dec")
        axs[2].set_ylabel("distance")
```

```python
#Arbitrary Close Object
orb = Orbit.from_name("Betelgeuse")
t = np.linspace(0,1,num=1001)*u.Myr
orb.integrate(t, MWPotential)
```

```python
#Make lists
GI = []
GG = []
AI = []
AG = []
#Find positions throughout orbit
for i in range(len(t)):
    #Initial conditions
    galpy_icrs = galpy_coord(orb, i, "icrs")
    galpy_galcen = galpy_coord(orb, i, "galcen")
    
    #Convert using astropy 
    astro_icrs = astro_to_icrs(galpy_galcen[0], galpy_galcen[1], galpy_galcen[2])
    astro_galcen = astro_to_galcen(galpy_icrs[0], galpy_icrs[1], galpy_icrs[2])
    
    #Append to lists 
    GI.append(galpy_icrs)
    GG.append(galpy_galcen)
    AI.append(astro_icrs)
    AG.append(astro_galcen)
    
GI = np.array(GI)
GG = np.array(GG)
AI = np.array(AI)
AG = np.array(AG)
```

```python
print(GI[:,0])
```

```python
#Plot values
fig, axs = plt.subplots(2,3)
#Top layer will be icrs
plot(axs[0], GI, t, "icrs", "purple")
plot(axs[0], AI, t, "icrs", "orange")
#Bottom layer will be galcen
plot(axs[1], GG, t, "galcen", "purple")
plot(axs[1], AG, t, "galcen", "orange")
plt.tight_layout()
```

```python
resid = np.abs(AG - GG)

fig, axs = plt.subplots(2,3)
plot(axs[0], GI, t, "icrs", "purple")
plot(axs[1], resid, t, "galcen", "red")
plt.tight_layout()
```

```python
resid = np.abs(AI - GI)

fig, axs = plt.subplots(2,3)
plot(axs[0], GG, t, "galcen", "purple")
plot(axs[1], resid, t, "icrs", "red")
plt.tight_layout()
```

This suggests that while astropy and galpy are the same in converting ra and dec into galactocentric coordinates, they tend to differ a lot on converting galactocentric coordinates back into ra and dec.

```python
S1 = []
S2 = []
C1 = []
C2 = []

for i in range(len(t)):
    #The canon
    c = galpy_coord(orb2, i , "icrs")
    
    #These two should be good
    s1 = galpy_coord(orb2, i , "galcen")
    s2 = astro_to_galcen(c[0],c[1],c[2])
    
    print(f"{s1}\n{s2}")
    
    #This shouldn't work
    convert1 = astro_to_icrs(s1[0], s1[1], s1[2])
    
    #This should work
    convert2 = astro_to_icrs(s2[0], s2[1], s2[2])
    
    S1.append(s1)
    S2.append(s2)
    C1.append(convert1)
    C2.append(convert2)
    C.append(c)
    
S1 = np.array(S1)
S2 = np.array(S2)
C1 = np.array(C1)
C2 = np.array(C2)
C = np.array(C)
```

```python
fig, axes = plt.subplots(3,3)
plot(axes[0], S1, t, "galcen", "purple")
plot(axes[0], S2, t, "galcen", "orange")
plot(axes[1], C, t, "icrs", "gold")
plot(axes[1], C2, t, "icrs", "orange")
plot(axes[2], C1, t, "icrs", "orange")
plt.tight_layout()
```
